```python
import numpy as np
import pandas as pd
import os
from pathlib import Path
from pdf2image import convert_from_path
from tqdm.notebook import tqdm
```

```python
#Actual code for manifest
poppler_path = "/Users/jimenagonzalez/miniforge3/envs/main/bin"
path = '/Users/jimenagonzalez/research/DSPL/Hybrid_SL_Modeling/Traditional_Modeling/OSG/' #Path with the modeling output from OSG

#To save the Zooniverse files
path_data = '/Users/jimenagonzalez/research/DSPL/Hybrid_SL_Modeling/Zooniverse/Data/'


folders = [f for f in os.listdir(path) if os.path.isdir(os.path.join(path, f))]
rows_model, rows_mask, rows_corner, rows_oneset = [], [], [], []

for folder in tqdm(folders):
    files = os.listdir(path + folder)
    for file in files:
        if not file.endswith('_corner.pdf'):
            continue
        
        coadd_id = int(file.split('_')[0])
        #rows_model.append({'OBJECT_COADD_ID': coadd_id, 'Image': f'{coadd_id}_model.png'})
        #rows_mask.append({'OBJECT_COADD_ID': coadd_id, 'Image': f'{coadd_id}_completeMasks.png'})
        #rows_corner.append({'OBJECT_COADD_ID': coadd_id, 'Image': f'{coadd_id}_corner.png'})
        rows_oneset.append({'OBJECT_COADD_ID': coadd_id, 'image_task1': f'{coadd_id}_completeMasks.png', 'image_task2': f'{coadd_id}_model.png',
                           'image_task3': f'{coadd_id}_corner.png'})

        #print(path, folder, file, coadd_id)
        # paths with the pdf files, and path with the new png files
        pdfpath_model = path + folder + '/' + str(coadd_id) + '_model.pdf' #I will change 'output_modeling N' ('folder') to just the coadd_id
        pdfpath_completeMasks = path + folder + '/' + str(coadd_id) + '_completeMasks.pdf'
        pdfpath_corner = path + folder + '/' + str(coadd_id) + '_corner.pdf'
        
        #newpath_model = path_data + 'Model/' + str(coadd_id) + '_model.png'
        #newpath_completeMasks = path_data + 'CompleteMasks/' + str(coadd_id) + '_completeMasks.png'
        #newpath_corner = path_data + 'Corner_plots/' + str(coadd_id) + '_corner.png'
        newpath_oneset = path_data + 'Oneset/' + str(coadd_id) 

        
        #Converting pdf to png:
        pages = convert_from_path(pdfpath_model, dpi=150, poppler_path=poppler_path)
        #pages[0].save(newpath_model, "PNG")
        pages[0].save(newpath_oneset + '_model.png', "PNG")

        pages = convert_from_path(pdfpath_completeMasks, dpi=150, poppler_path=poppler_path)
        #pages[0].save(newpath_completeMasks, "PNG")
        pages[0].save(newpath_oneset + '_completeMasks.png', "PNG")

        pages = convert_from_path(pdfpath_corner, dpi=20, poppler_path=poppler_path)
        #pages[0].save(newpath_corner, "PNG")
        pages[0].save(newpath_oneset + '_corner.png', "PNG")

#data_model = pd.DataFrame(rows_model)
#data_mask = pd.DataFrame(rows_mask)
#data_corner = pd.DataFrame(rows_corner)
data_oneset = pd.DataFrame(rows_oneset)

#data_model.to_csv(path_data + 'Model/manifest_model.csv', index=False)
#data_mask.to_csv(path_data + 'CompleteMasks/manifest_mask.csv', index=False)
#data_corner.to_csv(path_data + 'Corner_plots/manifest_corner.csv', index=False)
data_oneset.to_csv(path_data + 'Oneset/manifest.csv', index=False)
```

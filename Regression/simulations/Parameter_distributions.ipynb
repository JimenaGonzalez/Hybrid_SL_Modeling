```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table
```

```python
filename = 'Data/fit_files/1000_simu.fits'
hdu_list = fits.open(filename)
images= hdu_list[1].data 
data = hdu_list[2].data[:]
data = Table(data)
data = data.to_pandas()
    
print(len(data))
hdu_list.close()
```

```python
data.columns

params = ['EINSTEIN_RADIUS', 'e1_lens', 'e2_lens', 'center_x_lens', 'center_y_lens', 'R_sersic_lens_light', 
          'n_sersic_lens_light', 'e1_lens_light', 'e2_lens_light', 'R_sersic_source_light', 'n_sersic_source_light',
          'e1_source_light', 'e2_source_light', 'center_x_source', 'center_y_source']

for p in params:
    plt.figure(figsize=(7,4))
    plt.hist(data[p], density=True)
    plt.title(p)
    plt.xlabel(p)
    plt.ylabel("Probability density")
    plt.grid(alpha=0.3)
    plt.show()
```
